```python
import json
```

```python
file = open("1articles.json", encoding='utf-8')
```

```python
data = json.load(file)
```

```python
for i in range(len(data)):
    #create variables for the title and the definition title by removing all special characters
    try:
        #remove special chaaracters for definition title
        title = data[i]["title"]
        def_title = title.replace("/","_")
        def_title = def_title.replace("-","_")
        def_title = def_title.replace(":","_")
        def_title = def_title.replace("(","_")
        def_title = def_title.replace(")","_")
        def_title = def_title.replace("'","")
        def_title = def_title.replace("!","")
        def_title = def_title.replace(";","")
        def_title = def_title.replace(",","")
        def_title = def_title.replace("?","")
        def_title = def_title.replace(" ","_")
        url = data[i]["url"]
        url = url.replace("/article/","")
        #create html file and close it
        Func = open('1app.py', 'a')
        Func.write('@app.route("/'+f'{url}'+'\")\n'
         'def '+f'{def_title}'+'():\n'
         '\treturn render_template("'+f'{url}'+'.html")\n\n')
        Func.close()
    #skip if there is an error (some of the formatting is odd so this is a possibility)
    except:
        pass
```
